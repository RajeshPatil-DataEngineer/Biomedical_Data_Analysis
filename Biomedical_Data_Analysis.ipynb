```python
import pandas as pd

url = "http://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
names = ['preg','plas','pres','skin','insu','mass','pedi','age','class']
df:pd.DataFrame = pd.read_csv(url, names=names)
print(df)
```

```python
print("First 5 rows :", df.head())
print("\nDataFrame info", df.info())
print("\nDescription Statistics :", df.describe())
```

```python
print("Missing values per column: \n", df.isnull().sum())
```

```python
#Replace missing  values (if any). Let's replace 0's with NaN
import numpy as np #import numpy for NaN
print(df)
df[['preg','pres','skin','insu','mass']] = df[['preg','pres','skin','insu','mass']].replace(0,np.nan)
print(df)
print("Missing values per column after cleaning (replac 0 with NaN):\n", df.isnull().sum())
```

```python
#fill NaN with the mean value
df.fillna(df.mean(),inplace=True) #inplace = True modifies the DataFrame directly
print("\nDescriptive Statistics after Filling Missing Values")
print("\n___________________________________________________________\n")
print(df.describe())
```

```python
import matplotlib.pyplot as plt # impot pyplot for visual representation
print(df['mass'])
df.boxplot(column="mass")
plt.show() # display the plot
```

```python
#Clip the values (for demostration; be cautious with clipping)
mass_mean = df["mass"].mean()
mass_std = df["mass"].std()
print(f"Mean : {mass_mean}  Std : {mass_std}")

lower_bound = mass_mean - 3 * mass_std
upper_bound = mass_mean + 3 * mass_std
print(f"Lower Bound: {lower_bound} Upper Bound : {upper_bound}")

df["mass"] = df["mass"].clip(lower = lower_bound, upper = upper_bound)
df.boxplot(column = "mass")
plt.show()
```

Intuition
Mean = “Where is the data?”
Std = “How spread out is the data?”
📊 Example Comparison
Dataset A:

[20, 20, 20, 20]

Mean = 20
Std = 0 (no variation)
Dataset B:

[5, 20, 35, 50]

Mean = 27.5
Std = High (values spread out)



OTHER METHOD TO REMOVE THE OUTLIER.(FOR CHARTBOT)
IQR Method (Interquartile Range)

The IQR method uses quartiles to define a “normal range”:

Q1 (25th percentile) → lower quartile
Q3 (75th percentile) → upper quartile
IQR = Q3 - Q1

Then we define bounds:

Lower Bound = Q1 - 1.5 × IQR
Upper Bound = Q3 + 1.5 × IQR

```python
#The IQR method uses quartiles to define a “normal range”:
Q1 = df["mass"].quantile(0.25)
Q3 = df["mass"].quantile(0.75)
print(f"Q1 : {Q1} Q3 : {Q3}")

IQR = Q3 - Q1
print(f" IQR : {IQR}")

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Lower Bound: {lower_bound} Upper Bound : {upper_bound}")
```

```python
df["mass"] = df["mass"].clip(lower = lower_bound, upper = upper_bound)
df.boxplot(column = "mass")
plt.show()
```
