```python
import pandas as pd

url = "http://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
names = ['preg','plas','pres','skin','insu','mass','pedi','age','class']
df:pd.DataFrame = pd.read_csv(url, names=names)
print(df)
```

```python
print("First 5 rows :", df.head())
print("\nDataFrame info", df.info())
print("\nDescription Statistics :", df.describe())
```

Accessing Data in a DataFrame

```python
print(df)
print("Age column :",df['age']) #display all the data for the age column (vertical)
print("\nFirst row :", df.iloc[0]) # display all the data form the first row (horizental)
print("\nAge of the first row :", df.loc[0,'age']) # display only first row age column data (single column data)
```

Data cleaning and Preprocessing

Data issues, some of those imperfect ingredients
Missing values (need to find the way to fill them)
outliers (way-way outside the normal range)

```python
print("Missing values per column: \n", df.isnull().sum())
```

```python
#Replace missing  values (if any). Let's replace 0's with NaN
import numpy as np #import numpy for NaN
print(df)
df[['preg','pres','skin','insu','mass']] = df[['preg','pres','skin','insu','mass']].replace(0,np.nan)
print(df)
print("Missing values per column after cleaning (replac 0 with NaN):\n", df.isnull().sum())
```

```python
#fill NaN with the mean value
df.fillna(df.mean(),inplace=True) #inplace = True modifies the DataFrame directly
print("\nDescriptive Statistics after Filling Missing Values")
print("\n___________________________________________________________\n")
print(df.describe())
```
